```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Bengaluru_House_Data.csv")
```

```python
data.head(5)
```

```python
data
```

```python
data.drop(columns=['area_type','availability','society'],inplace=True)
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data['location'] = data['location'].apply(lambda x:x.strip())
```

```python
data.location.value_counts()
```

```python
location_stat = data.groupby('location')['location'].agg('count').sort_values(ascending=False)
```

```python
location_stat
```

```python
location_less_than_10_entries = location_stat[location_stat<=10]
```

```python
location_less_than_10_entries
```

```python
data['location'] = data['location'].apply(lambda x :'other' if x in location_less_than_10_entries else x) 
```

```python
data['location'].value_counts()
```

```python
data['size'].unique()
```

```python
data['bedrooms'] = data['size'].apply(lambda x:int(x.split(' ')[0]))
```

```python
data
```

```python
data.total_sqft.unique()
```

```python
def clean(sqft):
    tokens = sqft.split('-')
    if len(tokens)==2:
        return (float(tokens[0]) + float(tokens[1]))/2
    else:
        try:
            return float(sqft)
        except:
            return None
        
```

```python
data['total_sqft'] = data['total_sqft'].apply(clean)
```

```python
data.total_sqft.unique()
```

```python
data.describe()
```

```python
data.dropna(inplace=True)
```

```python
data.describe()
```

```python
data['sqft_per_bed'] = data['total_sqft']/data['bedrooms']
```

```python
data['sqft_per_bed'].describe()
```

```python
data2 = data[data['sqft_per_bed'] >= 300]
```

```python
data2
```

```python
data2['price_per_sq_ft'] = round(data2['price']*100000/data2['total_sqft'],2)
```

```python
data2['price_per_sq_ft'].describe()
```

```python
data2
```

```python
data3 = data2[data2['price_per_sq_ft'] >= 2000]
```

```python
data3
```

```python
data3.drop(columns = ['sqft_per_bed','price_per_sq_ft','size'],axis=0,inplace=True)
```

```python
data3
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
```

```python
col_trans = make_column_transformer((OneHotEncoder(sparse_output=False), ['location']), remainder='passthrough',force_int_remainder_cols=False)
```

```python
lr = LinearRegression()
scaler = StandardScaler()
model = make_pipeline(col_trans, scaler, lr)
```

```python
data_input = data3.drop(columns=['price'])
data_output = data3['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(data_input, data_output, test_size=0.2)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
input = pd.DataFrame([['Electronic City Phase II', 1500.0, 3.0, 2.0, 3]], columns=['location','total_sqft','bath','balcony','bedrooms'])
```

```python
model.predict(input)
```
